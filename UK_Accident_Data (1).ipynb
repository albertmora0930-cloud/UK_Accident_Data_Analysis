27. Selecting 6 to 7 district amongst the district area and apply specific analysis

```python
# 7 districts
district_data =  ['Birmingham', 'Leeds', 'Manchester', 'Liverpool', 'Sheffield', 'Bristol', 'Nottingham']

road_district = road[road['District Area'].isin(district_data)]
```

<h1>Compare Accident Severity (District vs Overall)</h1>

```python

plt.figure(figsize=(10,6))
sns.countplot(x='Accident_Severity', data=road, order=road['Accident_Severity'].value_counts().index, color='gray', alpha=0.5, label='Overall')
sns.countplot(x='Accident_Severity', data=road_district, order=road['Accident_Severity'].value_counts().index, color='red', alpha=0.6, label='Selected Districts')
plt.title("Accident Severity: Selected Districts vs Overall")
plt.legend()
plt.show()
```

<h1>Compare Casualties per Accident in Districts</h1>

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='District Area', y='Number_of_Casualties', data=road_district)
plt.title("Casualties per Accident in Selected Districts")
plt.xticks(rotation=45)
plt.show()
```

<h1>Compare Accident Counst by Districts</h1>

```python
road_district['District Area'].value_counts().plot(kind='bar', figsize=(8,5), color='skyblue')
plt.title("Number of Accidents per Selected District")
plt.ylabel("Accident Count")
plt.show()
```

<h1>Comapare Weather Conditions in Selected Districts</h1>

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Weather_Conditions', data=road_district, order=road['Weather_Conditions'].value_counts().index)
plt.title("Weather Conditions in Selected Districts")
plt.xticks(rotation=45)
plt.show()
```

<h1>Compare Accidents by Time of Day</h1>

```python
plt.figure(figsize=(10,6))
sns.histplot(road['Hour'], bins=24, color='gray', label='Overall', alpha=0.4)
sns.histplot(road_district['Hour'], bins=24, color='red', label='Selected Districts', alpha=0.6)
plt.title("Accidents by Hour: Selected Districts vs Overall")
plt.xlabel("Hour of Day")
plt.ylabel("Accident Count")
plt.legend()
plt.show()
```

Insight: Mosth accidents happen in daylight and involve cars, but serious crashes with more casualties often accur at night in bad weather, or on busy urban roads.
